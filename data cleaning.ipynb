```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Dhika Ramadhan Saputra\Data Analist\Portofolio\Klasifikasi Tingkat Kemiskinan di Indonesia\Klasifikasi Tingkat Kemiskinan di Indonesia.csv", sep=';')
df
```

```python
df.isnull().sum()
```

```python
df.drop(columns='Klasifikasi Kemiskinan', inplace=True)
```

```python
df.dropna(inplace=True)
df
```

```python
df['Umur Harapan Hidup (Tahun)'] = df['Umur Harapan Hidup (Tahun)'].str.replace(',','.')
df
```

```python
df['Umur Harapan Hidup (Tahun)'] = df['Umur Harapan Hidup (Tahun)'].astype(float)
df
```

```python
df['Umur Harapan Hidup (Tahun)'] = df['Umur Harapan Hidup (Tahun)'].round().astype(int)
df
```

```python
df['Rata-rata Lama Sekolah Penduduk 15+ (Tahun)'] = df['Rata-rata Lama Sekolah Penduduk 15+ (Tahun)'].str.replace(',','.')
df['Rata-rata Lama Sekolah Penduduk 15+ (Tahun)'] = df['Rata-rata Lama Sekolah Penduduk 15+ (Tahun)'].astype(float)
df['Rata-rata Lama Sekolah Penduduk 15+ (Tahun)'] = df['Rata-rata Lama Sekolah Penduduk 15+ (Tahun)'].round().astype(int)
df
```

```python
df['Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun)'] = df['Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun)'].astype(int)
df
```
